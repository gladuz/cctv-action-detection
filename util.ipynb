```python
import glob
import os

DATA_PATH = os.path.abspath('G:\내 드라이브\Project\RGBLab\ABB\Data')

all_files = list(glob.glob(os.path.join(DATA_PATH, '**', '*.mp4'), recursive=True))
```

```python
all_files
```

```python
# loop through all files and get the file name
for file in all_files:
    file_name = os.path.basename(file)
    
    # check if the file name contains '_resized.mp4_resized.mp4'
    if '_resized.mp4_resized.mp4' in file_name:
        
        # remove the file
        os.remove(file)
```

```python
import os
import pickle
rgb_features = pickle.load(open("/data/common/abb_project/processed_data/rgb_features.pkl", "rb"))
```

```python
rgb_normalized = {}
for k, v in rgb_features.items():
    filepath = "__".join(k.replace("\\", "__").split("__")[6:])
    rgb_normalized[filepath] = v
```

```python
print(len(rgb_normalized.keys()))
rgb_normalized.keys()
```

```python
import numpy as np
FLOW_DIR = "/data/common/abb_project/flow_features/"
flow_files = os.listdir(FLOW_DIR)
flow_normalized = {}
for fname in flow_files:
    flow_feature = np.load(os.path.join(FLOW_DIR, fname))
    flow_normalized[fname[:-4]] = flow_feature
```
