```python
from torch.utils.data import Dataset, DataLoader
import numpy as np
import os

class VideoFeatureDataset(Dataset):
    def __init__(self, root_folder):
        self.rgb_folder = os.path.join(root_folder, "rgb_kinetics_resnet50")
        self.flow_folder = os.path.join(root_folder, "flow_kinetics_bninception")
        self.label_folder = os.path.join(root_folder, "target_perframe")

        # Assuming all folders have matching file names
        self.file_names = os.listdir(self.rgb_folder)

        # Find the minimum length among all files
        self.min_length = float('inf')
        for file_name in self.file_names:
            rgb_path = os.path.join(self.rgb_folder, file_name)
            rgb_features = np.load(rgb_path)
            self.min_length = min(self.min_length, len(rgb_features))

    def __len__(self):
        return len(self.file_names)

    def __getitem__(self, idx):
        file_name = self.file_names[idx]

        rgb_path = os.path.join(self.rgb_folder, file_name)
        flow_path = os.path.join(self.flow_folder, file_name)
        label_path = os.path.join(self.label_folder, file_name)

        rgb_features = np.load(rgb_path)[:self.min_length]  # Truncate
        flow_features = np.load(flow_path)[:self.min_length]  # Truncate
        labels = np.load(label_path)[:self.min_length]  # Truncate

        combined_features = np.concatenate((rgb_features, flow_features), axis=1)
        return combined_features, labels

# Usage
# root_folder = "path/to/root_folder"
# dataset = VideoFeatureDataset(root_folder)
# loader = DataLoader(dataset, batch_size=2, shuffle=True)

```
